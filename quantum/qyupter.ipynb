```python
import random

import qnote
import qnote_util
```

```python
possible_states = [1, 2, 3, 4]
possible_bits = [[0,0], [1,0], [0,1], [1,1]]
```

```python
test_bits = []
test_bases = []
test_states = []
for i in range(8):
    idx = random.randint(0, 3)
    test_bits.append(possible_bits[idx][0])
    test_bases.append(possible_bits[idx][1])
    test_states.append(possible_states[idx])
qn = qnote.QNote("abc", test_states)
qc = qnote_util.gen_qnote_from_states(qn.states)
print(qn.serial)
print(qn.states)
print("ledger bits: ", test_bits)
print("ledger bases: ", test_bases)
print("verified: ", qnote_util.verify_qnote(qn, test_bits, test_bases))

qc.draw(output='mpl')
```
