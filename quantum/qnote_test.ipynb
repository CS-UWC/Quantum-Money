```python
import random

import qnote
import qnote_util
```

```python
possible_states = [1, 2, 3, 4]
possible_bits = [[0,0], [1,0], [0,1], [1,1]]
```

```python
idx = random.randint(0, 3)
test_bits = [0,0,1,0,1,0,0,1] # [possible_bits[idx][0]]
test_bases = [0,1,0,1,1,1,1,0] # [possible_bits[idx][1]]
test_states = [0,2,1,2,3,2,2,1] # [possible_states[idx]]
# 0 0 1 0 1 0 0 1
qn = qnote.QNote("abc", test_states)
qc = qnote_util.gen_qnote_from_states(qn.states)
print("QNote serial: ", qn.serial)
print("QNote states: ", qn.states)
print("Ledger bits: ", test_bits)
print("Ledger basis: ", test_bases)
print("Verified: ", qnote_util.verify_qnote(qn, test_bits, test_bases))

qc.draw(output='mpl')
```
